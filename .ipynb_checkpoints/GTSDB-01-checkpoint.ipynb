## Helper code

```python
def plot_rectangle(bbox, ax, class_name, edgecolor, confidence=None):
    xmin = bbox[0]
    ymin = bbox[1]
    xmax = bbox[2]
    ymax = bbox[3]
    left = xmin
    right = xmax
    top = ymin
    bot = ymax
    ax.add_patch(
        plt.Rectangle((left, top),
                      right-left,
                      bot-top, fill=False,
                      edgecolor=edgecolor, linewidth=3.5)
        )
    label = '{:s}'.format(class_name)
    label_pos_y = top-10
    if confidence:
        label += ' {0:.2f}'.format(confidence)
        label_pos_y = bot+20
    ax.text(left, label_pos_y,label,
            bbox=dict(facecolor=edgecolor, alpha=0.5),
            fontsize=14, color='white')
```

```python
def get_label_id(label_name):
    for category in categories:
        if category['name'] == label_name:
            return category['id']
```

## Detection

```python
min_score_thresh = 0.5
for image_path in TEST_IMAGE_PATHS:
    fig, ax = plt.subplots(figsize=(20, 20))
    image = Image.open(image_path)
    image_name = os.path.basename(image_path)
    width, height = image.size
    ax.imshow(image)
    
    image_np = load_image_into_numpy_array(image)
    image_np = image_np[:,:,::-1] # rgb -> bgr
    pred_results = tfnet.return_predict(image_np)
    
    print(f'pred results: {pred_results}')
    print()

    for idx, det in enumerate(pred_results):
        score = det['confidence']
        if score > min_score_thresh:
            bbox = det['topleft']['x'], det['topleft']['y'], det['bottomright']['x'], det['bottomright']['y']
            label = get_label_id(det['label'])
            plot_rectangle(bbox,ax,category_index[label]['name'],'red', score)
    plt.draw()
    fig.tight_layout()
    plt.axis('off')
    plt.show()
```

Although it seems beneficial to do the conversion from python command line instead of jupyter notebook. I tried all kinds of post-freezing graph surgeries (mainly input shape manipulations, even no-shape), to no avail.

My only other hope is that the darknet-based network will somehow convert to (first pb, which already succeeded, and then as a next step to) tflite. TFlite conversion hangs with "Here is a list of operators for which you will need custom implementations: ExtractImagePatches." This operation is currently not supported on tflite (and I read on a dev forum that it is not on the planned whitelist). If I somehow overcomme this hinderance, maybe the conversion will succeed (there are others who seem to convert YOLOv2 without problems: https://www.tooploox.com/blog/card-detection-using-yolo-on-android). As it turns out, they probably used an older version of darkflow (https://github.com/thtrieu/darkflow), which converted the yolov2 model to pb without using the ExtractImagePatches tf op. My intuition is proven correct as I find this post:

https://stackoverflow.com/questions/50061895/porting-darkflow-tensorflow-model-into-tensorflow-android-camera-detection-demo

"_No Opkernel means there is no implementation for the hardware running this .pb. To resolve this, look at class reorg of ./net/ops/convolution.py. It has two methods \_forward and forward. The current default option is using forward, which has extract_image_patches - a built-in method of tensorflow._

_Swap the names of two methods and you will be using my manual implementation, which should have no problem with Opkernel implementation._

Ref: https://github.com/thtrieu/darkflow/issues/56 https://github.com/thtrieu/darkflow/blob/master/darkflow/net/ops/convolution.py"

```python
graph_def_file = "/home/viktor/workspace/00_models/GTSDB_darknet_YOLOv2/built_graph/yolo_v2.pb"
input_arrays, output_arrays = ["input"], ["output"]

converter = tf.contrib.lite.TFLiteConverter.from_frozen_graph(graph_def_file, input_arrays, output_arrays)
tflite_model = converter.convert()
open("/home/viktor/Dropbox/00_VARINN/201904_TUV/gtsdb_yolov2.tflite", "wb").write(tflite_model)
```

```python
# Detector was trained on 1360 x 800 px GTSDB pics, is used to and requres the squeeze to 608x608
# this corresponds to a 17:10 ratio, which is close to 16:9 HD video, but not quite
ORIGINAL_W = 1360
ORIGINAL_H = 800
YOLOV2_WH = 608
YOLOV2_CELLSIZE = YOLOV2_WH / 19.
wscale = YOLOV2_WH / ORIGINAL_W
hscale = YOLOV2_WH / ORIGINAL_H

interpreter = tf.contrib.lite.Interpreter(model_path=f"/home/viktor/workspace/00_models/201904_TUV/gtsdb_yolov2.tflite")
interpreter.allocate_tensors()

# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
# print(input_details)
# print(output_details)

# Test model on random input data.
input_shape = input_details[0]['shape']

img = tf.keras.preprocessing.image.load_img('/home/viktor/workspace/00_datasets/test_images/02.jpg',
                                            target_size=(YOLOV2_WH, YOLOV2_WH))
x = tf.keras.preprocessing.image.img_to_array(img)/255.
# x = x[:,:,::-1] # rgb -> bgr
x = np.expand_dims(x, axis=0)

# input_data = np.array(np.random.random_sample(input_shape), dtype=np.float32)
interpreter.set_tensor(input_details[0]['index'], x)
interpreter.invoke()
output_data = interpreter.get_tensor(output_details[0]['index'])
# print(output_data.shape)
# uppleft = output_data[0, 0, 0]
# for i in range(8):
#     print(f'p{i}: {[uppleft[8*j+i] for j in range(5)]}')
    
objconfs = output_data[0][:, :, [4, 12, 20, 28, 36]]
objectanchors = np.where(objconfs>0)

def sigmoid(x):
    return 1/(1+np.exp(-x))

# prediction tuple: (objconf, class, ulx, uly, lrx, lry) i, j, x, y, w, h | 1 unit is one cell length
anchors = [[0.22, 0.37], [0.31, 0.52], [0.42, 0.71], [0.61, 1.02], [0.96, 1.56]]
for anchor in zip(objectanchors[0], objectanchors[1], objectanchors[2]):
    celli = anchor[0]
    cellj = anchor[1]
    anchornum = anchor[2]
    anchordata = output_data[0][celli, cellj, 8*anchornum:8*(anchornum+1)]
    x = cellj + sigmoid(anchordata[0])
    y = celli + sigmoid(anchordata[1])
    w = anchors[anchornum][0] * np.exp(anchordata[2]) * 1.2
    h = anchors[anchornum][1] * np.exp(anchordata[3]) * 1.2
    tlx = round(YOLOV2_CELLSIZE*(x - w/2) / wscale)
    tly = round(YOLOV2_CELLSIZE*(y - h/2) / hscale)
    brx = round(YOLOV2_CELLSIZE*(x + w/2) / wscale)
    bry = round(YOLOV2_CELLSIZE*(y + h/2) / hscale)
    objconf = sigmoid(anchordata[4])
    clazz = np.argmax(anchordata[5:8])
    print(f'found object in cell {celli:2d},{cellj:2d} at anchor#{anchornum}: {100*objconf:5.2f}% class{clazz}@{sigmoid(anchordata[5:8][clazz]):.2f} '\
         +f'ul: ({tlx},{tly}) lr: ({brx},{bry}) size: ({brx-tlx},{bry-tly})')
```
